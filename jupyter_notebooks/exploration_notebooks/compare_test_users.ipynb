```python
def categorize_days(val):
    if (val > 4) and (val <= 8):
        return 'Early Morning'
    elif (val > 8) and (val <= 12):
        return 'Morning'
    elif (val > 12) and (val <= 16):
        return 'Afternoon'
    elif (val > 16) and (val <= 20):
        return 'Evening'
    elif (val > 20) and (val <= 24):
        return 'Night'
    elif val <= 4:
        return 'Late Night'
    
def calc_groups(x):
    if 0 <= x <= .25:
        return "Low"
    elif .25 < x <= .50:
        return "Medium"
    elif .50 < x <= .75:
        return "High"
    elif .75 < x <= 1.0:
        return "Very High"
    
def create_hour_num(x):
    if x[1] > 9:
        return float(str(x[0]) + "." + str(x[1]))
    else:
        return float(str(x[0]) + "." + "".join("0" + str(x[1])))
```

```python
#Lets simply extract year month date hour min sec from the time series as a separate feature.
############### Working with Time Series Data ############################
type(time_data_set["created_at"][0])
#Converting to timestamp format for observation
time_data_set["created_at"] = pd.to_datetime(time_data_set["created_at"])

#Exploring the datetime

daysofweek = {
    0: "Monday",
    1: "Tuesday",
    2: "Wednesday",
    3: "Thursday",
    4: "Friday",
    5: "Saturday",
    6: "Sunday"
}

time_data_set["year"] = time_data_set["created_at"].apply(lambda x: x.year)
time_data_set["month"] = time_data_set["created_at"].apply(lambda x: x.month)
time_data_set["date"] = time_data_set["created_at"].apply(lambda x: x.date())
time_data_set["hour"] = time_data_set["created_at"].apply(lambda x: x.hour)
time_data_set["minute"] = time_data_set["created_at"].apply(lambda x: x.minute)
time_data_set["day"] = time_data_set["created_at"].apply(lambda x: x.day)
time_data_set["day_of_record"] = time_data_set["created_at"].apply(lambda x: daysofweek[x.dayofweek])
time_data_set["hour_num"] = time_data_set[["hour", "minute"]].apply(create_hour_num, axis=1)
time_data_set["hour_bins"] = time_data_set["hour_num"].apply(categorize_days)
```

```python
def calc_low(x):
    if 0 <= x < 0.05:
        return "0-0.05"
    elif 0.05 < x <= 0.10:
        return "0.05-0.10"
    elif 0.10 < x <= 0.15:
        return "0.10-0.15"
    elif 0.15 < x <= 0.20:
        return "0.15-0.20"
    elif 0.20 < x <= 0.25:
        return "0.20-0.25"

def calc_medium(x):
    if 0.25 < x <= 0.3:
        return "0.25-0.3"
    elif 0.3 < x <= 0.35:
        return "0.3-0.35"
    elif 0.35 < x <= 0.4:
        return "0.35-0.4"
    elif 0.4 < x <= 0.45:
        return "0.4-0.45"
    elif 0.45 < x <= 0.5:
        return "0.45-0.5"

def calc_high(x):
    if 0.5 < x < 0.55:
        return "0.5-0.55"
    elif 0.55 < x <= 0.60:
        return "0.55-0.60"
    elif 0.60 < x <= 0.65:
        return "0.60-0.65"
    elif 0.65 < x <= 0.70:
        return "0.65-0.70"
    elif 0.70 < x <= 0.75:
        return "0.70-0.75"

def calc_vhigh(x):
    if 0.75 < x < 0.80:
        return "0.75-0.8"
    elif 0.8 < x <= 0.85:
        return "0.8-0.85"
    elif 0.85 < x <= 0.90:
        return "0.85-0.90"
    elif 0.90 < x <= 0.95:
        return "0.90-0.95"
    elif 0.95 < x <= 1.0:
        return "0.95-1.0"
```

```python
low["sub_s_02"] = low.s02.apply(calc_low)
medium["sub_s_02"] = medium.s02.apply(calc_medium)
high["sub_s_02"] = high.s02.apply(calc_high)
very_high["sub_s_02"] = very_high.s02.apply(calc_vhigh)
```

```python
def plot_figure(x_axis, data, test_users, axis_order, settings):
    figure, ax = plt.subplots(settings["row"], settings["col"], figsize=settings["fig_size"])
    count = 0
    for axs in ax.flatten():
        if(count >= len(test_users)):
            break
        sns.countplot(x=x_axis, 
                    data=data[data["user_id"] == test_users[count]], order = axis_order,
                    ax=axs,
                    #palette="Blues")
                    color="steelblue")
        axs.set(title="".join(settings["pre_content"] + str(test_users[count])))
#         axs.set(title="".join("Test User - " + str(test_users[count]) + " " + 
#                               "Observations " + str(len(data[data["user_id"] == test_users[count]]))))
        axs.set_ylim(settings["ylim"])
        axs.set_xlabel("groups")
        axs.set_ylabel("frequencies")
        count+=1
    figure.tight_layout()
    if settings["delaxes"] is not None:
          for cord in settings["delaxes"]:
                figure.delaxes(ax[cord[0]][cord[1]])
    return figure, ax
```
