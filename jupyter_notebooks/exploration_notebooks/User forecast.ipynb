```python
def get_min_index_diff(usr_x_ts_index, usr_y_seq):
    minDiff = 1231412412
    min = -1
    i = 0
    min_idx = -1
    
    n = len(usr_y_seq)
    
    srtd_usr_seq = np.sort(usr_y_seq)
    
    while i < n:
        diff = np.abs(float(usr_x_ts_index) - srtd_usr_seq[i])
        #print(diff)
        if diff < minDiff:
            minDiff = diff
            min = srtd_usr_seq[i]
            min_idx = i

        i+=1
        
   # print("The min index is {} and the min difference is {} and the min val is {}"
    #      .format(min_idx, minDiff, min))
    return min_idx, minDiff
```

```python
import numpy as np


class EMA:
    def __init__(self, X):
        self.shape = X.shape[0]

    # Can set a threshold to compute the closeness and fartherness day is 0.03 note that. 
    # N days can be the gap. A setting parameter
    def ema_similarity(self,x,y, t= 0.03):
        #Select the subsequence containing the time_index, s02, s03
        #x_sub_seq = x[:,1:4]
        #y_sub_seq = y[:, 1:4]
        dist = 0.0
        months_count = 0
        minDiffSum = 0.0
        
        #For each subsequence compute a similarity with y near the time index. 
        # Sum against each of these days and return an average.
        n1 = len(x)
        n2 = len(y)
        
        if n1 == n2:
            month_range = n1
        elif n1 > n2:
            month_range = n2
        elif n2 > n1:
            month_range = n1
            
        
        for month in range(0, month_range):
            months_count += 1
            #print(seq1[0])
            #Compare to get the index which is minimum against x and y subsequences
            
            for seq in x[month][0]:
                min_idx, minDiff = get_min_index_diff(seq[0], y[month][0][:,0])
                #Fetch the row

                if minDiff <= t:
                    y_seq = y[month][0][min_idx,:]
                    #Loudness difference as a absolute, a frobenius norm, including the day_index
                    s02_diff = np.linalg.norm(seq[0:2]-y_seq[0:2])
                    dist+= s02_diff

                #Distress difference as absolute
                #s03_diff = np.linalg.norm((seq1[2]-y_seq[2]))

                #Average between both the difference
                #diff_avg = (s02_diff + s03_diff)/2

                #Compute a similarity against x and y. Here, we use frobenious norm
                #dist+= s02_diff
                #minDiffSum+=minDiff
            
        return (dist/months_count) + (np.abs(n1 - n2)/12)
```

```python
import numpy as np
import matplotlib.pyplot as plt 

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold


fig, ax = plt.subplots(9, 6, figsize=(50, 40))
plot_row=0
plot_col=0
for user, ts in user_time_series_mean.items():
    if plot_col == 6:
        plot_col=0
        plot_row += 1
    df = pd.DataFrame(ts)
    df.columns = ["day", "day_index", "s02", "s03", "s04", "s05", "s06", "s07"]
    X = df[["day_index", "s02", "s04", "s05", "s06", "s07"]]
    y = df[["s03"]]
    
    x_train, x_test, y_train, y_test = train_test_split(X, y.to_numpy().ravel(), test_size=0.2, random_state=1221)
    user_best_params = {}
    degree_rmse = {}
    min_rmse, min_deg = 1e10, 0
    degrees = []
    avg_rmse_list = []
    kf = KFold(n_splits=5, shuffle=False)

    for i in range(1, 6):
        count = 0
        rmses = []
        for train_idx, test_idx in kf.split(x_train):
            count += 1
            train = x_train.iloc[train_idx]
            label_train = y_train[train_idx]
            test = x_train.iloc[test_idx]
            label_test = y_train[test_idx]

            # Train features with PolynomialFeatures
            poly_features = PolynomialFeatures(degree=i, include_bias=False)
            x_poly_train = poly_features.fit_transform(train)

            # Linear regression with poly fit
            poly_reg = LinearRegression()
            poly_reg.fit(x_poly_train, label_train)

            # test data
            x_poly_test = poly_features.fit_transform(test)
            poly_predict = poly_reg.predict(x_poly_test)
            poly_mse = mean_squared_error(label_test, poly_predict)
            poly_rmse = np.sqrt(poly_mse)
            rmses.append(poly_rmse)
        degrees.append(i)
        avg_rmse_list.append(np.mean(rmses))

    from pprint import pprint
    degree = degrees[np.where(avg_rmse_list == np.min(avg_rmse_list))[0][0]]
    avg_min_rmse = np.min(avg_rmse_list)
    pprint('Best degree {} with the average min RMSE {} over 5-folds for user- {} '.format(degree, avg_min_rmse, str(user)))
    
    best_params = dict()
    best_params["degree"] = degree
    best_params["avg_min_rmse"] = avg_min_rmse
    user_best_params[str(user)] = best_params
    
    
    sns.lineplot(x=degrees, y=avg_rmse_list, ax=ax[plot_row][plot_col])
    #ax[plot_row][plot_col].set_yscale('log')
    ax[plot_row][plot_col].set_xlabel('Degree')
    ax[plot_row][plot_col].set_ylabel('Average RMSE (Log)')
    ax[plot_row][plot_col].set(title='Best degree {},  Average min RMSE {} over 5-folds, User- {} '.format(degree, avg_min_rmse, str(user)))
    plot_col+=1
    
    
fig.tight_layout()
fig.delaxes(ax[8][3])
fig.delaxes(ax[8][4])
fig.delaxes(ax[8][5])
plt.show()
```
